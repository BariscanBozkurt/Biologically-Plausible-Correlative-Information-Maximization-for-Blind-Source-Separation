# Source Generation and Mixing Scenario

```python
sound_paths = '../../data/Sounds'


rate, man = wavfile.read(os.path.join(sound_paths,'MA02_04.wav'))
rate, woman = wavfile.read(os.path.join(sound_paths,'FA01_03.wav'))

man = man.astype('float32')
woman = woman.astype('float32')

man_max = np.max(man)
woman_max = np.max(woman)
man /= man_max
woman /= woman_max

shortest = min(len(man), len(woman))
woman = woman[:shortest]
man = man[:shortest]


np.random.seed(101)
noise = np.random.uniform(-1, 1, len(man))*0.1
S = np.stack((woman, man, noise))
A = np.random.uniform(-1, 1, (3, 3))
# NumberofSources = S.shape[0]
# A =ortho_group.rvs(NumberofSources)
X = np.dot(A, S)


# SNRlevel = 30
# X, NoisePart = addWGN(X, SNRlevel, return_noise = True)

# SNRinp = 10 * np.log10(
#     np.sum(np.mean((X - NoisePart) ** 2, axis=1))
#     / np.sum(np.mean(NoisePart**2, axis=1))
# )
# print("The following is the mixture matrix A")
# display_matrix(A)
# print("Input SNR is : {}".format(SNRinp))
```

# Visualize Generated Sources and Mixtures

```python
subplot_1D_signals(
    S[:, 0:20], title="Original Signals", figsize=(15.2, 9), colorcode=None
)
subplot_1D_signals(
    X[:, 0:20], title="Mixture Signals", figsize=(15, 18), colorcode=None
)
```

# Algorithm Hyperparameter Selection and Weight Initialization

```python
lambday = 1 - 1e-2 / 10
lambdae = 1 - 1e-2 / 10
s_dim = S.shape[0]
x_dim = X.shape[0]

# Inverse output covariance
By = 5 * np.eye(s_dim)
# Inverse error covariance
Be = 1000 * np.eye(s_dim)

debug_iteration_point = 16000
model = OnlineCorInfoMaxICA(
    s_dim=s_dim,
    x_dim=x_dim,
    muW=1e-3,
    lambday=lambday,
    lambdae=lambdae,
    By=By,
    Be=Be,
    neural_OUTPUT_COMP_TOL=1e-6,
    set_ground_truth=True,
    S=S,
    A=A,
)
```

# Run CorInfoMax Algorithm on Mixture Signals

```python
with Timer() as t:
    model.fit_batch_ica(
        X=X,
        n_epochs=10,
        neural_dynamic_iterations=500,
        plot_in_jupyter=True,
        neural_lr_start=0.01,
        neural_lr_stop=0.001,
        debug_iteration_point=debug_iteration_point,
        shuffle=True,
    )
print("Algorithm took %f sec." % (t.interval))
```

# Visualize SINR Convergence 

```python
mpl.rcParams["xtick.labelsize"] = 18
mpl.rcParams["ytick.labelsize"] = 18
plot_convergence_plot(
    model.SIR_list,
    xlabel="Number of Iterations / {}".format(debug_iteration_point),
    ylabel="SINR (dB)",
    title="SINR Convergence Plot",
    colorcode=None,
    linewidth=1.8,
)

print("Final SINR: {}".format(np.array(model.SIR_list[-1])))
```

# Calculate Resulting Component SNRs and Overall SINR

```python
Wf = model.compute_overall_mapping(return_mapping=True)
Y = Wf @ X
Y_ = signed_and_permutation_corrected_sources(S, Y)
coef_ = ((Y_ * S).sum(axis=1) / (Y_ * Y_).sum(axis=1)).reshape(-1, 1)
Y_ = coef_ * Y_

print("Component SNR Values : {}\n".format(snr_jit(S, Y_)))

SINR = 10 * np.log10(CalculateSINRjit(Y_, S)[0])

print("Overall SINR : {}".format(SINR))
```
